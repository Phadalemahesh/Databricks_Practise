```python
from pyspark.sql import *
from pyspark.sql.functions import col, explode
from pyspark.sql.functions import concat_ws


df = spark.read.option("multiLine", True).format('json').load('/Volumes/data_mart_load/raw/raw_storage/Source_A/Raw_Data.json')

df = df.withColumn('email', col('profile.email')) \
        .withColumn('firstname', col('profile.firstName')) \
        .withColumn('lastname', col('profile.lastName')) \
        .withColumn('MultipleRoles', explode(col('roles'))) \
        .withColumn('pk',concat_ws('|',col('firstname'),col('lastname'),col('MultipleRoles')))

df2 = df.drop('preferences','profile','roles')

df2.write.format('delta').option('mergeSchema',True).mode('overwrite').saveAsTable('data_mart_load.raw.source_a_raw_data')
```
